```python
web_page = pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')
```

```python
sp = web_page[0]
```

```python
spunique = sp['Symbol'].unique()
spunique
```

```python
spenergy = sp[sp['GICS Sector'] == 'Energy']
```

```python
start_date = datetime.datetime(2021,1,25)
end_date = datetime.datetime.today()
```

```python
stock_price = pd.DataFrame()
```

```python
df = yf.download('API', start_date, end_date)
```

```python
df
```

```python
for i in spunique:
  df = yf.download(i, start_date, end_date)
  df['Symbol'] = i
  stock_price = stock_price.append(df)
  print(df.head())
```

```python
stock_price
```

```python
gg = stock_price[stock_price['Symbol'] == 'GL']

fig = go.Figure(data=[go.Candlestick(x=gg.index,
                open=gg['Open'], high=gg['High'],
                low=gg['Low'], close=gg['Close'])
                     ])

fig.show()
```
